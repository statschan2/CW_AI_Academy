## Line Chart

```python
data_url = ('https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-mean-temp.csv')
data = pd.read_csv(data_url)
data.head(3)
```

```python
data = data.set_index('Month')
data.head(3)
```

```python
plt.style.use("dark_background")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure(figsize = (15, 5))

ax = fig.add_subplot(1, 1, 1)
ax.plot(data.index, data['Temperature'], label = data.columns[0])
ax.set_title('월별 평균 온도', fontsize = 20)
ax.set_xlabel('Month', fontsize = 15)
ax.set_ylabel('Temp', fontsize = 15)

x_ticks = range(0, len(data.index), 6)
x_labels = [data.index[i] for i in x_ticks]
ax.set_xticks(x_ticks, x_labels, rotation = 45)

plt.grid(False)
plt.legend(loc = 'lower left', fontsize = 15)
plt.show()
```

## Scatter Plot
X축과 Y축에 수량이나 크기 등을 대응시켜서 적합한 점에 데잍터를 표시하는 그래프로 실제 데이터의 분포를 확인할 때 많이 활용된다.

```python
import numpy as np
```

```python
plt.style.use("dark_background")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)

fig = plt.figure(figsize = (5,5))

ax = fig.add_subplot(1,1,1)
ax.scatter(x, y)

plt.show()
```

### 예제 2

```python
data_url = ('https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/')
raw_data = os.path.join(data_url, 'anime_master.csv')
data = pd.read_csv(raw_data)
data.head(3)
```

```python
types = data['type'].unique()

plt.style.use('default')

fig = plt.figure()

ax = fig.add_subplot(1,1,1)
for i in types :
    x = data.loc[data['type'] == i, 'members']
    y = data.loc[data['type'] == i, 'rating']
    ax.scatter(x, y, alpha = 0.5, label = i)
    
plt.show()
```

### 예제 3

```python
from sklearn.datasets import load_iris
```

```python
raw_data = load_iris()
target = raw_data.target
features = raw_data.data
feature_names = raw_data.feature_names
```

```python
plt.style.use("default")
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure(figsize = (15,5))

ax1 = fig.add_subplot(1,2,1)
ax1.scatter(features[:,0], features[:,1], c = target)
ax1.set_title('Sepal Length / Sepal Width')
ax1.set_xlabel(feature_names[0], fontsize = 15)
ax1.set_ylabel(feature_names[1], fontsize = 15)

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(features[:,2], features[:,3], c = target)
ax2.set_title('Petal Length / Petal Width')
ax2.set_xlabel(feature_names[2], fontsize = 15)
ax2.set_ylabel(feature_names[3], fontsize = 15)

plt.show()
```

## Histogram
