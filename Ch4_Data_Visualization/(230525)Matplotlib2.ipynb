## Figure

### 기본 설정

```python
x = [0,1]
y = [0,1]

fig = plt.figure(figsize = (3,3) ,
                 dpi = 1000 ,          # dpi 는 사이즈 커기고 화질도 좋아짐 (해상도 조절)
                 facecolor = 'yellow', # facecolor 는 그래프 밖 색채우기 
                 edgecolor = 'cyan',   # 테두리 색
                 linewidth = 10,       # 테두리 두께
                 frameon = True        # 테두리 표시 여부
                 )

ax = fig.add_subplot(111 , facecolor = 'white', # add subplot 에 있는 facecolor  = 그래프 안 색채우기
                     xlim = [0,10],             # x 축 0~10 까지 설정
                     ylim = [0,10000])          # y 축 0~10000 까지 설정              
ax.plot(x, y)

plt.show()
```

### 배경색 설정

```python
x = [0,1]; x1 = [0,5,9]; x2 =  [0,0.6]; x3 =  [0.4, 0.4 ,0.6]
y = [0,1]; y1 = [9000,4000,7000]; y2 =  [1,1];y3 =  [0.0, 0.2, 0.5]

fig = plt.figure(figsize = (8,5) ,
                 dpi = 100 ,         
                 facecolor = 'yellow',   
                 edgecolor = 'cyan',      
                 linewidth = 10,     
                 frameon = True            
                 )

ax1 = fig.add_subplot(221 , facecolor = 'lightpink')
ax1.plot(x, y)

ax2 = fig.add_subplot(222 , facecolor = 'lavender', 
                      xlim = [0,10],        
                      ylim = [0,10000])
ax2.plot(x1, y1)

ax3 = fig.add_subplot(223 , facecolor = 'skyblue')
ax3.plot(x2 , y2)

ax4 = fig.add_subplot(224 , facecolor = 'turquoise')
ax4.plot(x3 , y3)

plt.show()
```

## Line Plot

### 그래프 선의 색과 굵기

```python
x = np.array([0,1])
y = [0,1]

fig = plt.figure(figsize = (9,3))

ax = fig.add_subplot(111, xlim = (-0.2, 4)) 

ax.plot(x, y, linewidth = 1 , color = 'b' , label = "linewidth=1(alpha=1)")
ax.plot(x + 0.6 , y , linewidth = 3 , color = 'g' , alpha = 0.5,  label = "linewidth=2(alpha=0.5)")
ax.plot(x + 1.2 , y , linewidth = 10 , color = 'r' , alpha = 0.7, label = "linewidth=10(alpha=0.7)")

plt.legend()
plt.show()
```

### 그래프 선의 끝 모양

```python
x = [0.1, 1]
y = np.array([0, 0])
 
fig = plt.figure(figsize = (6,3))

ax = fig.add_subplot(111 , xlim = (0, 1.7) , ylim = (-0.2, 1.2))

ax.plot(x, y + 0.6, linewidth = 12, solid_capstyle = 'butt', color = 'crimson', label = "butt")  
ax.plot(x, y + 0.2, linewidth = 12, solid_capstyle = 'round', color = 'lightgray', label = "round")  
ax.plot(x, y + 0.9, linewidth = 12, solid_capstyle = 'projecting', color = 'cyan', label = "projecting")

plt.legend(loc = 4)
plt.show()
```

### 그래프 선의 결합점 모양

```python
y = [0, 1.0, 0.2, 1.2]
x = np.array([1, 2, 3, 4])  

fig = plt.figure(figsize = (6, 3))

ax = fig.add_subplot(111, xlim = (0, 4.5))

ax.plot(x, y, linewidth = 2, solid_joinstyle = "miter", label = "Miter" )
ax.plot(x + 0.1, y, linewidth = 2, solid_joinstyle = "round", label = "Round" )
ax.plot(x + 0.2, y, linewidth = 2, solid_joinstyle = "bevel", label = "Bevel" )

ax.legend()
plt.show()
```

### 그래프 선의 종류
* 'solid'  :  -
* 'dashed'  :  --
* 'dotted'  :  ·

```python
x = np.array([0,1])
y = np.array([1.9,1.9])

fig = plt.figure(figsize = (6, 3))

ax = fig.add_subplot(111, xlim = (-0.1, 1.7), ylim = (0,2))

ax.plot(x, y, linewidth = 5, linestyle = 'solid', solid_joinstyle = "miter", solid_capstyle = 'butt', color = 'blue', label = "solid")
ax.plot(x, y - 0.1, linewidth = 5, linestyle = '-', solid_joinstyle = "miter", solid_capstyle = 'butt', color = 'blue', label = "-")

ax.plot(x, y - 0.4, linewidth = 5, linestyle = 'dashed', solid_joinstyle = "miter", solid_capstyle = 'butt', color = 'crimson', label = "dashed")
ax.plot(x, y - 0.6, linewidth = 5, linestyle = '--', solid_joinstyle = "miter", solid_capstyle = 'butt', color = 'crimson', label = "--")

ax.plot(x, y - 0.8, linewidth = 5, linestyle = 'dotted', solid_joinstyle = "miter", solid_capstyle = 'butt', color = 'cyan', label = "dotted")
ax.plot(x, y - 1, linewidth = 5, linestyle = ':', solid_joinstyle = "miter", solid_capstyle = 'butt', color = 'cyan', label = ":")

plt.legend()
plt.show()
```

## Scatter Plot

### 점의 색과 투명도

```python
x = np.arange(0, 6 , 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()

ax = fig.add_subplot(111)

ax.scatter(x, y, label = "sin", color = "lightblue", edgecolor = "blue" )
ax.scatter(x, z, label = "cos", color = "lightblue", edgecolor = "blue", alpha = 0.3 )

plt.legend(framealpha = 0.5 , loc = 'lower right')
plt.show()
```

### Ex 2

```python
from sklearn.datasets import load_iris 

iris = load_iris()
X = iris["data"]
df = pd.DataFrame(X , columns = iris.feature_names)

colors = ['lightblue', 'lightgreen', 'lightpink',"red"]

plt.style.use("classic")
plt.figure()
boxplot = plt.boxplot(df, labels = df.columns, notch = True, patch_artist = True)

for patch, color in zip(boxplot['boxes'], colors) :
    patch.set_facecolor(color)
plt.xticks(rotation=15, color="blue")
   
plt.show()
```

# 3. 추가 설정

## 수평선, 수직선

```python
r = np.random.rand(10)

plt.figure(figsize=(10,5))

plt.plot(r)
plt.axhline(0.75, color="red") 
plt.axvline(6 , color="green")

plt.show()
```

```python
r = np.random.rand(10)

plt.figure(figsize=(10,5))

plt.plot(r)
plt.axhline(0.75, xmin = 0.1, xmax = 0.5, color = "red")
plt.axvline(6,  ymin = 0.2, ymax = 0.7, color = "green")

plt.show()
```

## 부분 색칠

```python
r = np.random.rand(10)

plt.figure(figsize = (10,5))

plt.plot(r)
plt.axhline(0.75, color = "red") 
plt.axvline(6, color = "green")
plt.axhspan(0.25, 0.5, alpha = 0.3, color = 'orange', xmin = 0.6, xmax = 0.8)
plt.axvspan(2, 4, alpha = 0.3)

plt.show()
```

## 특정 공간 채우기

```python
r = np.random.rand(10)

plt.figure(figsize = (10, 5))

plt.plot(r)
plt.axhline(0.75, color = "red") 
plt.axvline(6, color = "green")
plt.axhspan(0.25, 0.5, alpha = 0.3, color = 'orange', xmin = 0.6, xmax = 0.8, hatch = "*", fill = False) 
plt.axvspan(2, 4, alpha = 0.3)

plt.show()
```

## 선 아래 특정 영역 채우기

```python
r1 = np.random.rand(10)
r2 = np.random.rand(10)
 
plt.figure(figsize=(5,2))

plt.plot(r1)
plt.plot(r2)
plt.axhline(0.75, color = "red") 
plt.axvline(6, color = "green")
plt.axhspan(0.25, 0.5, alpha = 0.3, color = 'orange', xmin = 0.6, xmax = 0.8, hatch = "*", fill = False) 
plt.axvspan(2, 4, alpha = 0.3) 
plt.fill_between(np.arange(2,7), r1[2:7])

plt.show()
```

## 그래프 사이 공간 채우기

```python
r1 = np.random.rand(10)
r2 = np.random.rand(10)
 
plt.figure(figsize=(5,2))

plt.plot(r1)
plt.plot(r2)
plt.axhline(0.75, color = "red") 
plt.axvline(6, color = "green")
plt.axhspan(0.25, 0.5, alpha = 0.3, color = 'orange', xmin = 0.6, xmax = 0.8, hatch = "*", fill = False) 
plt.axvspan(2, 4, alpha = 0.3) 
plt.fill_between(np.arange(2,7), r1[2:7], r2[2:7], color = "red", alpha = 0.4)

plt.show()
```
